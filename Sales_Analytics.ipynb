# Data Visualization

## Sales Trends Over Time

Group the data by month and analyze the total sales for each month.
Create a line plot or bar chart to visualize the sales trends over time.

```python
#df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].head()
```

Group the data by month and calculate total sales

```python
Sales_M = df.groupby('Month')['Sales'].sum().reset_index()
```

Methode 1 : Plotting sales trends over time (line plot)

Note : range(1, 13) --> Assuming months are from 1 to 12

```python
plt.figure(figsize = (8 , 4))
plt.plot(Sales_M['Month'] , Sales_M['Sales'] , marker = 'o' , linestyle = '-' , color = 'g')
plt.title('Sales Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

Methode 2 : Using Seaborn for a bar chart

```python
plt.figure(figsize = (8 ,4))
sns.barplot(x = 'Month' , y = 'Sales' , data = Sales_M , hue = 'Month', palette = 'viridis' , legend = False)
plt.title('Sales Trends Over Time ')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.show()
```

## Best-Selling Products

Identify the products with the highest quantity ordered.

Create a bar chart to visualize the top-selling products.

```python
Top_Product_S = df.groupby('Product')['Quantity Ordered'].sum().reset_index()
Top_Product_S.head()
```

Sort the products by quantity in descending order to get the top sellers

Top_Product_S = Top_Product_S.sort_values(by = 'Quantity Ordered' , ascending = False)
Top_Product_S.head()

```python
top_n = 10
Top_Product_S = Top_Product_S.head(top_n)
```

The resulting bar chart will show the top-selling products based on the quantity ordered

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Quantity Ordered', y='Product', hue= 'Product' ,data=Top_Product_S, palette='plasma' , legend = False )
plt.title('Top Selling Products')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
plt.show()
```

## Revenue Metrics

Calculate total sales .
Analyze sales performance in different cities.
Create visualizations (bar charts)

```python
Total_S_City = df.groupby('City')['Sales'].sum().reset_index()
Total_S_City.head(10)
```

Visualize total sales per city with a bar chart

```python
plt.figure(figsize = (8 , 4))
sns.barplot(x ='Sales' , y='City', data = Total_S_City , hue = 'City' ,palette = 'viridis' , legend = False )
plt.title('Total Sales Per City')
plt.xlabel('Total Sales ($)')
plt.ylabel('City')
plt.show()
```

## Hourly Analysis

Analyze sales patterns based on the hour of the day.
Identify peak hours for sales.

```python
Total_S_Hour = df.groupby('Hour')['Sales'].sum().reset_index()
Total_S_Hour.head()
```

Visualize hourly sales patterns with a line plot

```python
plt.figure(figsize = (8 , 4))
sns.lineplot(x = 'Hour', y = 'Sales' , data = Total_S_Hour , marker = 'o' , color ='purple')
plt.title('Hourly Sales Analysis')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```

```python
Peak_Hour = Total_S_Hour[Total_S_Hour['Sales'] == Total_S_Hour['Sales'].max()]
print(f"Peak Hours for Sales : {list(Peak_Hour['Hour'])}")
```
